```python
# custom marker styles for different types of amenities
fig, ax = plt.subplots(1, figsize=(10, 8))

# Plot the base layer (all amenities without specific styling)
gdf_amenities.plot(ax=ax, color='lightgrey')
color_column = 'amenity'
marker_styles = {'restaurant': ('H', 'red'), 'parking': ('D', 'blue')}
for typ, (marker, color) in marker_styles.items():
    subset = gdf_amenities[gdf_amenities[color_column] == typ]
    subset.plot(ax=ax,marker=marker, color=color, markersize=50, label=typ)
    
# #x and y axis labels
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Add a title
plt.title('Map of amenities')

# Show the plot
plt.show()
```

```python
gdf_road = gpd.read_file(r'C:/Users/Yasir/Downloads/Self Learning Courses/Geopandas Work/road.geojson')
gdf_boundary = gpd.read_file(r'C:/Users/Yasir/Downloads/Self Learning Courses/Geopandas Work/boundry.geojson')

# Set up the plot
fig, ax = plt.subplots(figsize=(12, 10))

# Plot the line and polygon on the same axis
gdf_road.plot(ax=ax, color='gray', label='Roads')
gdf_boundary.boundary.plot(ax=ax, color='blue', linewidth=1, label=' Sector Boundary')

# Add a legend
ax.legend()

# Add a title
plt.title('Roads and Boundary')

# Set axis titles
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Show the plot
plt.show()
```

```python
# clip roads to the Sector boundary of the polygon

# Clip the roads to the polygon using geopandas
clipped_road = gpd.clip(gdf_road, gdf_boundary)

# Set up the plot
fig, ax = plt.subplots(figsize=(12, 10))

# Plot the clipped layer
clipped_road.plot(ax=ax, color='gray', label='Clipped Roads')

# Plot the polygon on the same axis
gdf_boundary.boundary.plot(ax=ax, color='blue', linewidth=1, label='Admin Map Boundary')

# Add a legend
ax.legend()

# Add a title
plt.title('Clipped Roads')

# Set axis titles
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Show the plot
plt.show()
```
